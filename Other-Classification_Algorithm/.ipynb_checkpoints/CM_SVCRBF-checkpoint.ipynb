```python
#!/usr/bin/env python

import numpy as np

# Load training and testing data
X_train = np.loadtxt('input_classification/X_train.csv', delimiter=',', skiprows=1)
X_test = np.loadtxt('input_classification/X_test.csv', delimiter=',', skiprows=1)
y_train = np.loadtxt('input_classification/y_train.csv', delimiter=',', skiprows=1)[:, 1]
```

```python
from sklearn.linear_model import LogisticRegression 

logisticRegr = LogisticRegression()

X_train_sub = X_train[:,:50] #indeed we get a better accuracy with a subset of features in logistic regression

logisticRegr.fit(X_train_sub, y_train)
print(logisticRegr)

predictions = logisticRegr.predict_proba(X_test[:,:50])
print(np.mean(predictions, axis=0))
print(predictions)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# PCA for logistic regression

n_components=4

scaler = StandardScaler() 
x = scaler.fit_transform(X_train)

pca = PCA(n_components)
principalComponents = pca.fit_transform(x)

#pca test
test_scaled = scaler.fit_transform(X_test)
pca_test = PCA(n_components)
principalComponents_test = pca_test.fit_transform(test_scaled)

print(principalComponents)

#for i in range(X_train[0].size):
#    np.append(X_train[:,i])



```

```python
fig = plt.figure()
plt.plot(principalComponents[c1,0],  principalComponents[c1,1], 'ro')
plt.plot(principalComponents[c2,0],  principalComponents[c2,1], 'bo')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(principalComponents[c1,0],  principalComponents[c1,1], principalComponents[c1,2], c='r', marker='o') #1st cluster

ax.scatter(principalComponents[c2,0],  principalComponents[c2,1], principalComponents[c2,2], c='b', marker='o') #1st cluster

ax.view_init(30, 280)

plt.xlabel("PC 1")
plt.ylabel("PC 2")
```

```python
# lets try L1 regulurisation and cross validation (accuracy with PCA=0.778)

from sklearn.linear_model import LogisticRegressionCV

logisticRegr = LogisticRegressionCV(cv=10, penalty='l1', solver='saga',  tol=1e-6, max_iter=int(1e6))


logisticRegr.fit(principalComponents, y_train)
print(logisticRegr)

predictions = logisticRegr.predict(principalComponents_test)
print(predictions)

```

```python
# In order to use SVM successfully we need to scale our data,
# this removes the mean from each feature and makes them each unit variance.
# this makes the individual features look like standard normally distributed data.
# it is essential for SVM, otherwise features with large variance dominate the objective function 

scaler = StandardScaler() 
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV


# find the best parameters for the SVM model, class_weight is new feature to test
# code here adapted from scikit-learn documentation examples
def svc_param_selection(X, y, nfolds):
    Cs = [0.001, 0.01, 0.1, 1, 1.1, 1.2]
    gammas = [0.0001, 0.001, 0.005, 0.01, 0.02, 0.03]
    param_grid = {'C': Cs, 'gamma' : gammas}
    grid_search = GridSearchCV(svm.SVC(kernel='rbf', class_weight='balanced', tol=1e-6), param_grid, cv=nfolds, scoring='accuracy')
    grid_search.fit(X, y)
    grid_search.best_params_ 
    
    print("Grid scores on development set:")
    print()
    means = grid_search.cv_results_['mean_test_score']
    stds = grid_search.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, grid_search.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
            % (mean, std * 2, params))
    
    return(grid_search.best_params_)

print(svc_param_selection(X_train_scaled, y_train, 3))
```

```python
#fit and classify the svm model with the best params
clf = svm.SVC(kernel='rbf', C = 1, gamma=0.005, class_weight='balanced', tol=1e-6)
clf.fit(X_train_scaled, y_train)
predictions = clf.predict(X_test_scaled)
```

```python
print(predictions[0:10])
```

```python
# Arrange answer in two columns. First column (with header "Id") is an
# enumeration from 0 to n-1, where n is the number of test points. Second
# column (with header "EpiOrStroma" is the predictions.
test_header = "Id,EpiOrStroma"
n_points = X_test.shape[0]
y_pred_pp = np.ones((n_points, 2))
y_pred_pp[:, 0] = range(n_points)
y_pred_pp[:, 1] = predictions
np.savetxt('CM_SVCRBF.csv', y_pred_pp, fmt='%d', delimiter=",",
           header=test_header, comments="")

# Note: fmt='%d' denotes that all values should be formatted as integers which
# is appropriate for classification. For regression, where the second column
# should be floating point, use fmt='%d,%f'.
```

```python
# split the data to evaluate specicifity and sensitivity
from sklearn.metrics import classification_report,confusion_matrix,f1_score
from sklearn.model_selection import train_test_split
X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(X_train_scaled, y_train, test_size=0.3, shuffle=False)

clf = svm.SVC(kernel='rbf', C = 1, gamma=0.005, class_weight='balanced', tol=1e-6)
clf.fit(X_train_split, y_train_split)
y_pred = clf.predict(X_test_split)
print(classification_report(y_test_split, y_pred))
```
